```python
import sqlite3
import pandas as pd
import glob
import os

# Connect (creates if not exists)
conn = sqlite3.connect("stocks.db")
cursor = conn.cursor()

# Create table (with Dividends + Stock Splits)
cursor.execute("""
CREATE TABLE IF NOT EXISTS prices (
    symbol TEXT,
    date TEXT,
    open REAL,
    high REAL,
    low REAL,
    close REAL,
    volume REAL,
    dividends REAL,
    stock_splits REAL,
    PRIMARY KEY (symbol, date)
)
""")

# Loop through all CSVs
for file in glob.glob("stocks_data/*.csv"):
    symbol = os.path.basename(file).replace(".csv", "")
    print(f"Inserting {symbol}...")

    # Read CSV
    df = pd.read_csv(file)

    # Make sure Date exists
    if "Date" not in df.columns:
        df.reset_index(inplace=True)
        df.rename(columns={"index": "Date"}, inplace=True)

    # Standardize column names
    df = df.rename(columns={
        "Date": "date",
        "Open": "open",
        "High": "high",
        "Low": "low",
        "Close": "close",
        "Volume": "volume",
        "Dividends": "dividends",
        "Stock Splits": "stock_splits"
    })

    # Add symbol column
    df["symbol"] = symbol

    # Reorder columns
    df = df[[
        "symbol", "date", "open", "high", "low", "close",
        "volume", "dividends", "stock_splits"
    ]]

    # Insert into DB
    df.to_sql("prices", conn, if_exists="append", index=False)

conn.commit()
conn.close()

print("✅ All stocks inserted into stocks.db")
```
